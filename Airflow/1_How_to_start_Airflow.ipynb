## Airflow 시작하기

```python
# 설치
!pip install apache-airflow
```

```python
# db init
!airflow db init
```

```python
# run web server
# 해당 명령어는 계속 실행되고 있어야 하는데
# 그러면 이 뒤의 명령어들이 jupyter에서 돌아가지 않으므로
# 터미널에서 돌려놓고 있겠다.
!airflow webserver -p 8080
```

```python
# create airflow user
!airflow users create --role Admin --username admin --email admin --firstname admin --lastname admin --password admin
```

------------
Airflow를 실행하고 User를 만들어 localhost:8080으로 접속하면 다음과 같은 화면을 띄우는 데에 성공한다.


화면의 상단을 보면 WORNING이 떠 있는 것을 볼 수 있다.

```python
# 첫번째 WORNING은 scheduler가 없으면 DAG를 업데이트 수 없다는 뜻으로
# 다음 명령어를 실행하고 리프레시 해보면 없어진다.
# 한번 실행시켰을 때는 잘 됐는데 다시 하면 에러가 반복적으로 일어난다.
# 근데 localhost:8080에 들어가보면 WORNING없어져있다.
!airflow scheduler
```

```python
# dags 커맨드로 dag들을 다룰 수 있다.
!airflow dags list
```

```python
# dag에 포함된 task 조회
!airflow tasks list example_xcom
```

```python
# dag trigger
!airflow dags trigger -e 2023-01-01 example_xcom
```

trigger한 dag는 위와 같이 DagRun인스턴스가 생성되었다는 메시지가 뜨고
localhost:8080에 접속해보면 다음과 같이 표시가 생긴다.

## 다시 시작하기
설치 및 사용법을 익힌 후 airflow를 다시 실행시켰더니 다음과 같은 에러가 나오며 실행되지 않는다.

airflow.exceptions.AirflowException: The webserver is already running under PID 5046.




프로세스가 이미 돌고있다는것 같아서 localhost:8080에 접속해봤지만 접속되지 않는다.

```python
# 프로세스가 진짜 돌고있는지 확인해봤다
# 확인해보니 위의 캡쳐 아래부분에서 5046에서 PID가 running중이라는 안내와 같이 
# 5046에서 프로세스가 돌고있었다(저 프로세스와 웹을 띄워주는 건 별개인가보다)
!ps -aef | grep airflow
```

```python
# 일단 죽여서 다시 시작해보면 성공했다.
!kill 5046
!airflow webserver -p 8080
```

```python
# 다시 시작하면 잊지 말고 scheduler도 실행시키자
!airflow scheduler
```
